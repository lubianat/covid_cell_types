```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import re

req = Request("https://www.covid19cellatlas.org/index.healthy.html")
html_page = urlopen(req)

soup = BeautifulSoup(html_page, "lxml")

links = []
for link in soup.findAll('a'):
    try:          
        classes = link.get("class")
        if 'btn-primary' in classes:
            if ".cxg" in link.get("href"):
                links.append(link.get("href"))
            else:
                links.append("https://www.covid19cellatlas.org" + link.get("href"))
    except:
        continue
        
print(links)
```

```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
import time


def get_cell_types(url):
    """
    Scrapes cell types from COVID-19 visualization portal
    """

    driver = webdriver.Firefox()
    driver.get(url)

    all_cell_types = []
    try:
        print("Waiting 20 seconds")
        time.sleep(20)
        
        cookies="button.bp3-intent-primary:nth-child(1) > span:nth-child(1)"
        driver.find_element_by_css_selector(cookies).click()
        
        
        cell_type_names = ["annotation:category-expand", 
                           "broad_celltype:category-expand", 
                           "BroadCellType:category-expand", 
                           "CellType:category-expand", 
                           "celltype:category-expand", 
                           "cell_type:category-expand"]
        
        for i in driver.find_elements_by_xpath("//span[@data-testid]"):
            title = i.get_attribute("data-testid")
            print(title)

            if title in cell_type_names:
                try:
                    i.click()
                except:
                    print("oops")

        for i in driver.find_elements_by_xpath("//span[@aria-label]"):
            label = i.get_attribute("aria-label")
            all_cell_types.append(label)
        driver.quit()
        print(all_cell_types)
        return(all_cell_types)

    except:
        return("failed")
    
    

all_datasets = {}
for url in links:
    print(f"Running for {url}")
    cell_types =  get_cell_types(url)
    all_datasets[url] = set(cell_types)
```
